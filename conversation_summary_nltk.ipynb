# Using nltk to summarize a text

### 1. Import the necessary dependencies.

```python
import nltk
from nltk. corpus import stopwords
from nltk. tokenize import word_tokenize, sent_tokenize
import pandas as pd
```

### 2. Import the csv file, extract the text as a new line separated concatenated string, then tokenize the text.

After importing the csv file as a dataframe, I want to extract the text column as a concatenation of strings.
Once I have the text in the desired format, I will use 'word_tokenize' to identify all the words in the text and stopwords to identify the stopwords, which are words that are very frequent and don't convey specific meaning.

```python
# Import the csv and extract the text column as new line separated concatenated string.
transcript_df = pd.read_csv('transcript.csv')
transcript_df.columns = ['person_name','text']
text = transcript_df.text.str.cat(sep='\n')


# Set stopwords to the english language and use word_tokenize to identify all the
# words in the text.

stopWords = set(stopwords.words("english"))
words = word_tokenize(text)
```

### 3. Create frequency dictionaries to keep the score values of words and sentences in the text.

First, I will create a dictionary called freqTable that stores the frequency of each word in the text; the function also filters all the english stopwords that were defined previously in the list 'stopWords': as explained before, these words don't contribute to the meaning of the text so that they will be excluded.
As a next step, I will use 'sent_tokenize' to divide the text in sentences and create another for loop to define the score value for each sentence, defined as the sum of scores of each word that creates the sentence.

```python

#Creating a frequency table to keep the score of each word

freqTable = dict()

for word in words:
    word = word.lower ()
    if word in stopWords:
          continue
    if word in freqTable:
          freqTable[word] += 1
    else:
          freqTable[word] = 1



# Tokenizing the sentences and creating a dictionary to keep the score of each 
# sentence

sentences = sent_tokenize(text)
sentenceValue = dict()


for sentence in sentences:
    for word, freq in freqTable.items():
        if word in sentence.lower():
            if sentence in sentenceValue:
                sentenceValue[sentence] += freq
            else:
                sentenceValue[sentence] = freq
```

### 4. Define the average value score of a sentence in the text, then create a text summary that only contains sentences with a value score that is 1.2 times higher than the average.

```python
# Define sumValues as the sum of all sentence values
sumValues = 0
for sentence in sentenceValue:
    sumValues += sentenceValue[ sentence]

# Define average value of a sentence from the original text
average = int (sumValues / len (sentenceValue))

# Storing sentences into our summary.
Summary = ''
for sentence in sentences:
    if (sentence in sentenceValue) and (sentenceValue[sentence] > (1.2 * average)) :
        Summary += "\n"+ sentence
```

### 5. Compare the original text and the summarized one.

```python
print(text)
```

```python
print(Summary)
```

The summarized text is nearly 3 times shorter than the original and that is not a bad result.
